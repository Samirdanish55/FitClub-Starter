```python
# Simple Student Grade Calculator
# -*- coding: utf-8 -*-

def get_marks(prompt="Enter your marks (0-100): "):
    while True:
        s = input(prompt).strip()
        try:
            marks = float(s)
        except ValueError:
            print("Please enter a valid number.")
            continue
        if 0 <= marks <= 100:
            return marks
        print("Marks must be between 0 and 100. Try again.")

def grade_and_message(marks):
    if marks >= 90:
        return "A", "Excellent work! Keep it up!"
    elif marks >= 80:
        return "B", "Great job! You're doing really well!"
    elif marks >= 70:
        return "C", "Good effort! Keep improving!"
    elif marks >= 60:
        return "D", "You passed, but you can do even better!"
    else:
        return "F", "Don't worry! Keep practicing and you'll improve."

def main():
    marks = get_marks()
    grade, message = grade_and_message(marks)
    print("\n--- Result ---")
    print(f"Marks: {marks}")
    print(f"Grade: {grade}")
    print(f"Message: {message}")

if __name__ == "__main__":
    main()
```
