```python
import pandas as pd
import numpy as np

# -----------------------------
# Generate Sample Sales Dataset
# -----------------------------

np.random.seed(42)   # Ensures reproducible results

products = ["Laptop", "Phone", "Tablet", "Headphones", "Smartwatch"]
quantities = np.random.randint(1, 20, size=60)
prices = np.random.choice([200, 500, 800, 1200, 1500], size=60)
product_choices = np.random.choice(products, size=60)

df = pd.DataFrame({
    "Product": product_choices,
    "Quantity": quantities,
    "Price": prices
})

# Calculate Sales
df["Sales"] = df["Quantity"] * df["Price"]

# -----------------------------
# Analysis
# -----------------------------

# 1. Total Sales
total_sales = df["Sales"].sum()

# 2. Best-Selling Product (highest total revenue)
product_sales = df.groupby("Product")["Sales"].sum()
best_product = product_sales.idxmax()
best_product_value = product_sales.max()

# 3. Basic Sales Report
print("===== SALES REPORT =====")
print(df.head(10))  # Show first 10 rows only
print("\nTotal Sales:", total_sales)
print(f"Best-Selling Product: {best_product} (Sales = {best_product_value})")

print("\n=== Product Revenue Breakdown ===")
print(product_sales)
```
